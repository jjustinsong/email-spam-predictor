```python
original_training_data = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Convert the emails to lowercase as the first step of text processing.
original_training_data['email'] = original_training_data['email'].str.lower()
test['email'] = test['email'].str.lower()

original_training_data.head()
```

```python
original_training_data = original_training_data.fillna('')
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(original_training_data, test_size = 0.1, random_state = 42)

train = train.reset_index(drop = True)
```

```python
def words_in_texts(words, texts):
    indicator_array = []
    for i in texts:
        arr = []
        for j in words:
            if j in i:
                arr.append(1)
            else:
                arr.append(0)
        indicator_array.append(arr)
    return np.asarray(indicator_array)
```

```python
def num_words(text):
    return len(text.split())

def cap_prop(text):
    count = 0
    for c in text:
        if c.isupper():
            count += 1
    if count > 0:
        return count/len(text)
    else:
        return 0

def re_or_fw(text):
    match = text.str.extract(r"(Fw:|Re:)")
    return match[0].notna().astype(int)

def special_char(text):
    match = text.str.extractall(r"([^\w ])")
    counts = match.groupby(level=0).size()
    return counts.reindex(text.index, fill_value=0)

train['e_num_words'] = train['email'].astype(str).apply(num_words)
train['s_num_words'] = train['subject'].astype(str).apply(num_words)
train['e_cap_prop'] = train['email'].astype(str).apply(cap_prop)
train['s_cap_prop'] = train['subject'].astype(str).apply(cap_prop)
train['re_or_fw'] = re_or_fw(train['subject'])
train['s_char'] = train['subject'].astype(str).apply(len)
train['e_char'] = train['email'].astype(str).apply(len)
train['e_special'] = special_char(train['email'])
train['s_special'] = special_char(train['subject'])
train
```

```python
sns.stripplot(x=train['re_or_fw'], y=train['s_cap_prop'], hue=train['spam'])
plt.title('Capital letter proportion in subject vs. Whether or not email is a reply/forward')
plt.xlabel('Reply/Forward')
plt.ylabel('Capital letter percentage')
```

```python
words = ['html', 'body', 'font', 'click', 'href', 'please', 'offer']
words_df = pd.DataFrame(words_in_texts(words, train['email']))
words_df['spam'] = train['spam']
words_df_melt = words_df.melt('spam')
sns.barplot(x=words_df_melt['variable'], y=words_df_melt['value'], hue=words_df_melt['spam']).set(xticklabels=words)
plt.xlabel('Words')
plt.ylabel('Proportion of emails')
plt.title('Words and the proportion of emails they are found in')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_curve, confusion_matrix
import re
from collections import Counter
```

```python
def click_count(text):
    words = text.split()
    count = Counter(words)
    return count['click']

def please_count(text):
    words = text.split()
    count = Counter(words)
    return count['please']

def offer_count(text):
    words = text.split()
    count = Counter(words)
    return count['offer']

click = train[train['spam'] == 1]['email'].apply(click_count)
please = train[train['spam'] == 1]['email'].apply(please_count)
offer = train[train['spam'] == 1]['email'].apply(offer_count)

count = 0
for i in offer:
    if i > 0:
        count += 1

print(count/len(please))
```

```python
words = ['html', 'body', 'font', 'click', 'href', 'please', 'offer']
word = pd.DataFrame(words_in_texts(words, train['email']))
word['id'] = train['id']
train = train.merge(word, on='id')
train
```

```python
train.columns = train.columns.astype(str)
X_train = train.iloc[:, 4:]
Y_train = train['spam']
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_train)
print(accuracy_score(Y_train, prediction))
```
